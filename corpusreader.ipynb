# damenconvlex

XMLCorpusReader + XMLCorpusView: https://nltk.googlecode.com/svn/trunk/doc/api/nltk.corpus.reader.xmldocs-module.html<br>
    
**(1)** als Plaintext in NLTK Korpus

```python
from nltk.corpus import XMLCorpusReader
corp = XMLCorpusReader('../../Knowledge Base/Korpora, Encyclopedia, Lexika/damenconvlex', 'DamenConvLex-1834-TEI.xml')
```

```python
corp.words()
```

**(2)** Abfrage mit XMLCorpusView

```python
from nltk.corpus.reader.xmldocs import XMLCorpusView
view = XMLCorpusView('../../Knowledge Base/Korpora, Encyclopedia, Lexika/damenconvlex/DamenConvLex-1834-TEI.xml','.*/text/body/div')
```

```python
print (view[2][0].text, '==', view[2][1].text)
```

**(3)** Abfrage mit lxml (= flexiblere Alternative)

```python
from lxml import etree
tree = etree.parse('../../Knowledge Base/Korpora, Encyclopedia, Lexika/damenconvlex/DamenConvLex-1834-TEI.xml')

for elt in tree.iter('{http://www.tei-c.org/ns/1.0}div'):
    if elt.attrib.get('type'):
        print (elt.get('n'))
```

# tigercorpus

for column types see:<br>
* http://www.nltk.org/_modules/nltk/corpus/reader/conll.html<br>
* http://www.ims.uni-stuttgart.de/forschung/ressourcen/werkzeuge/TIGERSearch/doc/html/TigerXML.html<br>
* https://stackoverflow.com/questions/20332762/pos-tagging-german-texts-using-nltk/23607200#23607200<br>
<br>

```python
from nltk.corpus import ConllCorpusReader
columns = ['ignore', 'words', 'ignore', 'ignore', 'pos']
corp = ConllCorpusReader('../../Knowledge Base/Korpora, Encyclopedia, Lexika/tigercorpus-2.2.conll09', 'tiger_release_aug07.corrected.16012013.conll09', columns, encoding='utf8')
```

```python
corp.words()
```

```python
corp.tagged_words()
```

```python
corp.tagged_sents()
```
